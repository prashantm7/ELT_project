````python
responseDict = { "page" : {} }
````

````python
responseDict
````

### Making request to delete records on loop

````python
%%time

for i in range(281):
  print(f'---- running cycle for #{i} --------')
  pageNum = str(i)
  currParams = ",".join(leadListDict['page'][pageNum])
  urlDelete = url.format(currParams)
  response = requests.request("DELETE", urlDelete, headers=myHeader)
  jsonResponse = json.loads(response.text)
  print('  --> saving response in dictionary <--  ')
  responseDict['page'][pageNum] = jsonResponse['data']
  time.sleep(2)
````

### saving responseDict as a json

````python
%%time

with open('deleteResponse_invoiceItems.json','w',encoding='utf-8') as f:
  f.write(json.dumps(responseDict,indent=4,default=str,sort_keys=True))

print('done')
````
